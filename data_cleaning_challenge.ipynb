# Importing Necessary Libraries

```python
# Data Manipulation
import pandas as pd

#ignore warning messages 
import warnings
warnings.filterwarnings('ignore')
```

# Lodaing Data

```python
data_import = pd.read_csv("data_cleaning_challenge.csv")
data_import
```

# Data Cleaning

```python
drop_extra = data_import.drop(columns=["Unnamed: 9", "Unnamed: 10"])
drop_extra
```

```python
drop_nans = drop_extra[drop_extra["Row Type"].notna()]
drop_nans
```

```python
column_values = []

counter = 0
for i in drop_nans["Row Type"]:
    if "first name" in i:
        counter += 1
    column_values.append(counter)
```

```python
iter_cols = drop_nans
iter_cols["Iteration"] = column_values
iter_cols
```

```python
drop_extra_column_name_rows = iter_cols[iter_cols["Row Type"] != "Row Type"]
drop_extra_column_name_rows
```

```python
name_dataframe = drop_extra_column_name_rows[drop_extra_column_name_rows["Row Type"].str.contains("first name")]
name_dataframe.drop(columns=["Speed1","Speed2","Electricity","Effort","Weight","Torque"], axis=1, inplace=True)
name_dataframe.rename(columns={
    "Row Type": "First Name",
    "Iter Number": "Last Name",
    "Power1": "Date"
}, inplace=True)
name_dataframe["First Name"] = name_dataframe["First Name"].str[12:]
name_dataframe["Last Name"] = name_dataframe["Last Name"].str[11:]
name_dataframe["Date"] = name_dataframe["Date"].str[6:]
```

```python
name_dataframe
```

```python
no_name_dataframe = drop_extra_column_name_rows[~drop_extra_column_name_rows["Row Type"].str.contains("first name")]
no_name_dataframe
```

```python
cleaned_df = pd.merge(left=name_dataframe, right=no_name_dataframe, how="inner", on="Iteration")
```

## Data Cleaned Successfully

# Save the Dataframe as csv file

```python
cleaned_df.to_csv("cleaned_data.csv", index=False)
```
